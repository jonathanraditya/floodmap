```python
#Insert all file to single array

rawScrapFile=[]
with os.scandir('../tides/test1') as entries:
    for entry in entries:
        #print(entry.name)
        rawScrapFile=rawScrapFile+[entry.name]
        
#print(rawScrapFile)

#Check and collect duplicate
noFilesDuplicate=[]
collectDuplicate=[]
for i in range(len(rawScrapFile)):
    if rawScrapFile[i] not in noFilesDuplicate:
        noFilesDuplicate=noFilesDuplicate+[rawScrapFile[i]]
    else :
        collectDuplicate=collectDuplicate+[rawScrapFile[i]]

print("Duplicate Files : ", collectDuplicate)
noFilesDuplicate.sort(reverse=False)
print(noFilesDuplicate)
```

```python
#Storing all crawled Coordinates in array

crawledCoordinates1=[]
crawledCoordinates2=[]
explodedDataArray=[]

def explodeArray(fileName, replaceParams, splitParams):
    replaces=fileName.replace(replaceParams, '')
    splitToArrays=replaces.split(splitParams)
    return splitToArrays

for i in range(len(noFilesDuplicate)):
    
    #Split name into array
    splitToArray=explodeArray(noFilesDuplicate[i], ']', '[')
    explodedDataArray=explodedDataArray+[splitToArray]
    #print(split)
    if splitToArray[3] not in crawledCoordinates1:
        crawledCoordinates1=crawledCoordinates1+[splitToArray[3]]
    else:
        crawledCoordinates2=crawledCoordinates2+[splitToArray[3]]

#pint(crawledCoordinates1)
#pint(explodedDataArray)

print(explodedDataArray[10][3])
print(len(crawledCoordinates1))

    
```

```python
#Storing MD5 for each crawled Coordinates

#Making Function of MD5 checksum
import hashlib
def md5(fname):
    hash_md5=hashlib.md5()
    with open(fname, 'rb') as f:
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()



def searchFile(coorName, explodedFileArray):
    #search full name from array
    #firstIndex=explodedFileArray.index(coorName)
    firstIndex=0
    secondIndex=0
    for i in range(len(explodedFileArray)):
        if explodedFileArray[i][3] == coorName:
            if firstIndex==0:
                firstIndex=i
            else:
                secondIndex=i
    return firstIndex, secondIndex

sR1, sR2=searchFile("95.01130555555555,-8.354685606060606", explodedDataArray)
print(sR1, sR2)
print(noFilesDuplicate[sR1],"\n" ,noFilesDuplicate[sR2])

fileName1='../tides/test1/'+noFilesDuplicate[sR1]
fileName2='../tides/test1/'+noFilesDuplicate[sR2]
with open(fileName1, 'r') as f:
    data=f.read()
    
print(md5(fileName1))
print(md5(fileName2))
    
#print(data)


    
```

```python
#Scan for file size anomaly
def fileSize(filePath):
    import os
    osStat=os.stat(filePath)
    return osStat.st_size

#Loop for each files
md5CC1=[]
md5CC2=[]
fileSize1=[]
fileSize2=[]

for i in range(len(crawledCoordinates1)):
    sResultIdx1, sResultIdx2 = searchFile(crawledCoordinates1[i], explodedDataArray)
    fileName1='../tides/test1/'+noFilesDuplicate[sResultIdx1]
    fileName2='../tides/test1/'+noFilesDuplicate[sResultIdx2]
    #insert MD5
    md5CC1=md5CC1+[md5(fileName1)]
    md5CC2=md5CC2+[md5(fileName2)]
    fileSize1=fileSize1+[fileSize(fileName1)]
    fileSize2=fileSize2+[fileSize(fileName2)]

#print(md5CC1)

#Check for files with different md5
differentMD5=[]
for i in range(len(md5CC1)):
    if md5CC1[i] != md5CC2[i]:
        differentMD5=differentMD5+[i]

#print(differentMD5)

#Comparing and filtering with standard file size
referenceFileSize=25000 #bytes

#print(fileSize1)
fileSizeAnomaly1=[]
fileSizeAnomaly2=[]
for i in range(len(fileSize1)):
    if fileSize1[i] >= referenceFileSize:
        fileSizeAnomaly1=fileSizeAnomaly1+[i]
    if fileSize2[i] >= referenceFileSize:
        fileSizeAnomaly2=fileSizeAnomaly2+[i]
        
print(fileSizeAnomaly1)
    
```

```python
#Cek irisan dari kedua anomali (md5 dan size)
indexToRetry=differentMD5.copy()
for i in range(len(fileSizeAnomaly1)):
    if fileSizeAnomaly1[i] not in indexToRetry:
        indexToRetry=indexToRetry+[fileSizeAnomaly1[i]]
for i in range(len(fileSizeAnomaly2)):
    if fileSizeAnomaly2[i] not in indexToRetry:
        indexToRetry=indexToRetry+[fileSizeAnomaly2[i]]

print(len(differentMD5),'\n', sep="")
print(len(fileSizeAnomaly1),'\n', sep="")
print(len(fileSizeAnomaly2),'\n', sep="")
print(len(indexToRetry))
print(indexToRetry)
```

```python
filePath='../tides/test1/[2][32][95.71912094017094,1.0481237373737375][hasil(1)-1602398717703].txt'
with open(filePath, 'r') as g:
    eliminate=g.read()
    
def replaceWithBlank(contents, identifier):
    return contents.replace(identifier, '')
def coordShortener(longCoord):
    return round(longCoord, 4)
def hourDateGenerator(year, month, day, numbOfDays):
    import datetime
    from datetime import time, timedelta
    initialDate=datetime.datetime(year, month, day)
    generateAllHours=24*numbOfDays
    resultArray=[]
    for i in range(generateAllHours):
        resultArray=resultArray+[initialDate+timedelta(hours=i)]
    return resultArray
    
#Header replace
removeHeader=' \n     Lat       Lon        yyyy-mm-dd hh:mm:ss (UTC)     z(m)\n \n'
eliminate=replaceWithBlank(eliminate, removeHeader)

testHDG=hourDateGenerator(2020,10,1, 15)

#Per-line replace


#eliminate=replaceWithBlank(eliminate, perLineReplaceParams)
for i in range(len(testHDG)):
    latitude=coordShortener(95.71912094017094)
    longitude=coordShortener(1.0481237373737375)
    #print(latitude, longitude)
    
    perLineReplaceParams='     '+str(longitude)+'   '+str(latitude)+'     '+str(testHDG[i])+'     '
    perLineReplaceParamsNegative='     '+str(longitude)+'   '+str(latitude)+'     '+str(testHDG[i])+'    '
    eliminate=replaceWithBlank(eliminate, perLineReplaceParams)
    eliminate=replaceWithBlank(eliminate, perLineReplaceParamsNegative)
    

#for i in range(len(testHDG)):
    #print(testHDG[i])


print(eliminate)
```
